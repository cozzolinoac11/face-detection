# Face detection

```python
import cv2

def face_detection(image):
  detectors = [ cv2.CascadeClassifier('face_detector.xml'),
                cv2.CascadeClassifier('lbpcascade_profileface.xml'),
                cv2.CascadeClassifier('haarcascade_profileface.xml')
              ]
  img = cv2.imread(image)
  for detector in detectors:
    faces = detector.detectMultiScale(img, 1.1, 4)
    if len(faces) > 0:
      for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.imwrite(image.split('.')[0]+"_face_detected.png", img)
        print('Successfully saved')
        return image.split('.')[0]+"_face_detected.png"
  img = cv2.flip(img, 1)
  for detector in detectors:
    faces = detector.detectMultiScale(img, 1.1, 4)
    if len(faces) > 0:
      for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        img = cv2.flip(img, 1)
        cv2.imwrite(image.split('.')[0]+"_face_detected.png", img)
        print('Successfully saved')
        return image.split('.')[0]+"_face_detected.png"
  print('No faces detected')
  return image
```
